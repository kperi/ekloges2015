```python
import numpy as np
import scipy as sp
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
data_2012 = pd.read_csv('data_2012.csv', sep='\t')
data_2013 = pd.read_csv('data_2013.csv', sep='\t')
data_2014 = pd.read_csv('data_2014.csv', sep='\t')
```

```python
data_2012.head(n=10)
data_2012 = data_2012.sort('Date')
```

```python
firms = set( data_2012['Polling Firm'])
```

```python
firms
```

```python
parties = ['ND','SYRIZA', 'PASOK/ Elia', 'ANEL', 'XA', 'DIMAR', 'KKE','Potami','Others']
dates = sorted( list(data_2012['Date']) )

def format_date(dt):
    parts = dt.split('_')
    return "%s-%s-%s" % (parts[1], parts[2] , parts[0])

dates = map(format_date, dates)
```

```python
plt.figure(num=None, figsize=(17, 14), dpi=80, facecolor='w', edgecolor='k')
for p in parties:
    df = data_2012[ [ p] ]
    plt.plot(df.values, label=p)
 
plt.ylabel(u"Percenage %")
plt.xlabel(u"poll day")
locs, labels = plt.xticks(range(len(dates)), dates, size='small')
plt.setp(labels, rotation=45)
plt.legend(loc='lower left')

plt.show()
```
